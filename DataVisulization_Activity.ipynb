SCATTERPLOT

```python
plot.scatter(football["Age"],football["Jersey Number"])
plot.title("Scatter plot for age and Jersey Number")
plot.xlabel("Jersey Number")
plot.ylabel("Age")
plot.show()
```

Bar chat

```python
plot.figure(figure=(8,5))
plot.bar(football["Name"],football["Age"])
plot.title(" Player name versus age in football dataset ")
plot.xlabel("Name")
plot.ylabel("Age")
plot.show()
```

Pie chart for barnds and market share

```python
brands=["Apple", "Samsung","Huawei","Xiaomi"]
market_share=[30,30,20,20]

plot.figure(figure=(7.7))
plot.pie(market_share,labels=brands,autopct='%1.1f%%',startangle=140)
plot.title("Smartphone market share")
plot.show()
```

Histogram

```python
data=np.random.normal(loc=20.0,scale=5.0,size=365)
#average temp 20, std deviation 5
plot.hist(data,bins=20,color='blue',alpha=0.7)
plot.title("Hostogram of Daily temperature Over One Year")
plot.xlabel("Temperature(degrees Celcius)")
plot.ylabel("Frequency")
plot.show()
```

Creation of a box plot for test_scores

```python
test_scores=np.random.normal(70,10,100)#Normal test_scores 
test_scores=np.append(test_scores,[40,40,90,100,100])
plot.boxplot(test_scores,vert=True,patch_artist=True,meanline=True,showmeans=True)
plot.title("Distribution of test_scores")
plot.ylabel("Test_scores")
plot.grid(True)
plot.show()
```
